These are just supplementary functions I added to the program.

`sub_strints()` -> just subtracts the two number strings and returns it back as a string

`equalize_lengths()` -> A small algorithm made to equalize the lengths of number strings that not the same length. Aka from recursively using the multiplication function. Takes the max length of the one of the two numbers. Another variable is multiplied by 2's to find how many leading zero's to add to make both strings equal length. The algorithm then returns both strings with the leading zeros added to the start of the numbers so they are of equal length

```python
def add_strints(x: str, y: str) -> str:
  """
    Add two nonnegative integer strings by converting to int.
    This method can be rewritten as a sum/carry adder for a
    single digit addition, pulling characters from the
    input strings. For simplicity now, we just convert the
    whole string to integer, do the addition, and then
    convert the number back to string.
  """
  return str(int(x) + int(y))

def sub_strints(x: str, y: str) -> str:
  """
    Subtract two nonnegative integers by converting them to integers and then returning the value as a string.

    Added this method as a supplementary function.
  """
  return str(int(x) - int(y))

def equalize_lengths(x: str, y: str) -> list[str]:
  """
    Pad the shorter string with leading zeros so that both strings have the same length.
  """
  n = max(len(x), len(y))
  p = 1 
  while p < n: 
    p *= 2

  return [x.zfill(p), y.zfill(p)]
```

`simple_recursive_multiplication`
Kept mostly the same structure as the given function. Modified it and removed any excess lines of code.

```python
def simple_recursive_multiplication(x: str, y: str) -> str:
  """
  Recursive multiplication for nonnegative integer strings.
  Assumptions:
    - len(x) == len(y)
    - len(x) is a power of two
    -  and y contain only digits
  Uses: 
    xy = ac*10^n + (ad+bc)*10^(n/2) + bd
  """
  # Equalize lengths of x, y
  padded = equalize_lengths(x,y)
  x = padded[0]
  y = padded[1]

  # Number of digits in x, y
  n = len(x)

  # Base case
  if n == 1:
      return str(int(x) * int(y))
  
  # Middle of x, y for splitting them in left/right halves
  m = n // 2

  # Divide x, y into left/right halves
  a = x[:m]
  b = x[m:]
  c = y[:m]
  d = y[m:]

  # karatsub's improvement
  # xy = ac10^2m + ((a+b)(c+d) - ac - bd)*10^m + bd

  # Compute ac and bd
  ac = simple_recursive_multiplication(a,c)
  bd = simple_recursive_multiplication(b,d)

  # Compute (a+b)(c+d)
  a_plus_b = add_strints(a,b)
  c_plus_d = add_strints(c,d)

  # ab * cd
  ab_cd = simple_recursive_multiplication(a_plus_b,c_plus_d)
  
  ad_plus_bc = sub_strints(sub_strints(ab_cd,ac),bd)

  #shift by powers of 10
  term1 = ac + ("0" * (2*m))
  term2 = ad_plus_bc + ("0" * (m))  

  return add_strints(term1, add_strints(term2, bd)) 
```
